```julia
using Plots
```

```julia
g = 9.8
vx0=20
ts = range(0,10,step=0.1)
ys = 1/2 * g.*ts.^2
xs = vx0.*ts
vs= round.(g.*ts,digits=3)
theta = range(0,2π,step=0.01)
r = 1 # 質点の半径
anim = Animation()
for i in 1:length(ts)
    plt=plot(xs[i] .+r*cos.(theta),ys[i].+r*sin.(theta),aspectratio=1,xlims=(-10,100),ylims=(-10,100),yaxis=:flip,xlabel="x",ylabel="y",linewidth=5)
    frame(anim,plt)
end
gif(anim,fps=15)
```

```julia
g = 9.8
vx0=20
ts = range(0,10,step=0.1)
ys = 1/2 * g.*ts.^2
xs = vx0.*ts
vs= round.(g.*ts,digits=3)
theta = range(0,2π,step=0.01)
r = 1 # 質点の半径
anim = Animation()
plot()
for i in 1:length(ts)
    plt=plot!(xs[i] .+r*cos.(theta),ys[i].+r*sin.(theta),aspectratio=1,xlims=(-10,100),ylims=(-10,100),yaxis=:flip,label="",linewidth=3,linecolor=:blue)
    #plt=plot!(xs[1:i],ys[1:i],label="",linewidth=5)
    plt=plot!(0 .+r*cos.(theta),ys[i].+r*sin.(theta),label="",linewidth=3,linecolor=:red)
    plt=plot!(xs[i] .+r*cos.(theta),0 .+r*sin.(theta),label="",linewidth=3,linecolor=:green)
    frame(anim,plt)
end
gif(anim,fps=20)
```
