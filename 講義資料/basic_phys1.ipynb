# 物理基礎 1編 1章　3節　落体の運動

## 自由落下

自由落下：物体が重力だけを受け,初速度$0$で鉛直下向きに落下する運動

重力加速度：自由落下の加速度



```julia
using Plots
```

```julia
g = 9.8
ts = range(0,10,step=0.1)
ys = 1/2 * g.*ts.^2
vs= round.(g.*ts,digits=3)
theta = range(0,2π,step=0.01)
r = 5 # 質点の半径
anim = Animation()
for i in 1:length(ts)
    plt=plot(0 .+r*cos.(theta),ys[i].+r*sin.(theta),aspectratio=1,xlims=(-50,50),ylims=(-10,100),yaxis=:flip,label="t=$(ts[i])",ylabel="y",linewidth=5)
    frame(anim,plt)
end
gif(anim,fps=15)
```

初速度$0$,加速度$g$の運動の速度は時間$t$に対して

$$
    v = g t
$$

である。
これをグラフにしてみると、

```julia
i=1
plot(0 .+r*cos.(theta),ys[i].+r*sin.(theta),aspectratio=1,xlims=(-50,50),ylims=(-10,100),yaxis=:flip,label="t=$(ts[i])",ylabel="y",linewidth=5)
savefig("basicphys1")
```

```julia
anim=Animation()
for i in 1:length(ts)
    plt=plot(ts[1:i],vs[1:i],xlabel="second",ylabel="m/s",xlims=(0,10),ylims=(0,100),label="t=$(ts[i])",linewidth=5)
    frame(anim,plt)
end
gif(anim,fps=15)
```

このようになる。

この運動の変位を考えたい。
$$
     変位 = 速度\times 時間
$$

であったことを思いだすと、ある時刻での変位は

```julia
anim=Animation()
for i in 1:length(ts)
    plt=plot(ts[1:i],vs[1:i],xlabel="second",ylabel="m/s",xlims=(0,10),ylims=(0,100),label="t=$(ts[i])",linewidth=5)
    plt = plot!(Shape([0,ts[i],ts[i]],[0,0,vs[i]]),alpha=0.3,label="")
    frame(anim,plt)
end
gif(anim,fps=15)
```

この図の影の部分の面積として表せる。

これは三角形の面積公式で計算することができて、
底辺を経過時間、高さを速度に対応させれば、

$$
    y  = \frac{1}{2} gt^2
$$

が得られる。


いま、速度と変位を時間で表す式

$$v = gt$$

$$y = \frac{1}{2} gt^2$$

が得られている。
この二つの式から$t$を消去することを考える。
すると、

$$v^2=2gy$$
が得られる。
これは、変位と速度を結びつける式と解せる。

## 鉛直投射

鉛直投げ下ろし：下向きに初速度を与えられた物体が、重力だけを受けて落下する運動

鉛直投げ上げ：上向に初速度を与えら得れた物体が、重力だけを受けて、上昇したあと落下する運動



### 鉛直投げ下ろし

鉛直投げ下ろしは自由落下と違い初速度がある。
この初速度を$v_0(>0)$とすれば、速度を時間で表す式は
$$v = v_0 + gt$$
と書くことができる。

これをグラフにしてみると、

```julia
anim = Animation()
g = 9.8
v0=20
ts = range(0,10,step=0.1)
ys = 1/2 * g.*ts.^2 + v0.*ts
vs= round.(g.*ts.+v0,digits=3) 
for i in 1:length(ts)
    plt=plot(ts[1:i],vs[1:i],xlabel="second",ylabel="m/s",xlims=(0,10),ylims=(0,100),label="t=$(ts[i])",linewidth=5)
    plt = plot!(Shape([0,ts[i],ts[i],0],[0,0,vs[i],20]),alpha=0.3,label="")
    frame(anim,plt)
end
gif(anim,fps=15)
```

のようになる。

自由落下のときと同じように、図の影の部分の面積がその時刻での変位を表している。
面積を計算すれば、変位を時間で表した式が得られるはずだ。
台形の面積とみなして計算しても良いし、長方形と三角形の和とみなして計算しても良い。
その結果は
$$y = v_0 t + \frac{1}{2} g t^2$$
である。

ここまでで、鉛直投げ下ろしについて、速度を時間で表す式、変位を時間で表す式がそれぞれ
$$v = v_0 + gt$$
$$y = v_0 t + \frac{1}{2} g t^2$$
と得られている。

この二つの式から時間を消去すれば、
$$v^2-v_0^2 = 2gy$$
が得られる。
これは変位と速度を結びつける式と解せる。

### 鉛直投げ上げ

鉛直投げ上げ：上向に初速度を与えら得れた物体が,重力だけを受けて,上昇したあと落下する運動

鉛直投げ上げ運動の様子は以下のようである。
自由落下・鉛直投げおろしの例と異なり、縦軸上向を正に取っていることに注意する。

```julia
g = 9.8
v0=20
ts = range(0,10,step=0.1)
ys = -1/2 * g.*ts.^2 + v0.*ts
vs= round.(-g.*ts.+v0,digits=3) 
anim = Animation()
theta = range(0,2π,step=0.01)
r = 5 # 質点の半径
for i in 1:length(ts)
    # plt=plot(0 .+r*cos.(theta),ys[i].+r*sin.(theta),aspectratio=1,xlims=(-50,50),ylims=(-10,100),yaxis=:flip,label="v = $(vs[i])")
    plt=plot(0 .+r*cos.(theta),ys[i].+r*sin.(theta),aspectratio=1,xlims=(-50,50),ylims=(-50,50),label="t=$(ts[i])",linewidth=5)
    frame(anim,plt)
end
gif(anim,fps=15)
```

いま、縦軸上向を正に取っている。
したがって、重力加速度は$-g$となる。
初速度を$v_0(>0)$とすれば、速度を時間で表した式は
$$v = v_0 -gt$$
とかける。
これをグラフにしてみると、

```julia
anim=Animation()
for i in 1:length(ts)
    plt=plot(ts[1:i],vs[1:i],xlabel="second",ylabel="m/s",xlims=(0,10),ylims=(-100,30),label="t=$(ts[i])",linewidth=5)
    plt = plot!(Shape([0,ts[i],ts[i],0],[0,0,vs[i],20]),alpha=0.3,label="")
    frame(anim,plt)
end
gif(anim,fps=15)
```

のようになる。
図の影の部分の面積がその時刻での変位を表している。
いま、速度(縦軸)が正の領域にある面積と負の領域にある面積がある。
$$変位 = 速度\times 時間$$
であったことを思い出すと、速度が正の領域の面積は正の方向への変位を表し、速度が負の領域の面積は負の方向への変位を表すと解せる。
運動のアニメーションと見比べると、正の面積を張り終わった時間に質点は最高点に到達し、直後負の面積を張り始めるとともに落下を始めることが確認できる。

ある時刻での面積を計算すれば、それがその時刻の変位である。
ただし、速度が正の領域の面積を正とし、速度が負の領域の面積を負として足し合わせる。
すると、
$$y = v_0t - \frac{1}{2}gt^2$$
が得られる。

自由落下、鉛直投げ下ろしの場合と同じように、
$$v = v_0 -gt$$
$$y = v_0t - \frac{1}{2}gt^2$$
から$t$を消去すれば、
$$v^2-v_0^2 =-2gy$$
が得られる。

## 水平投射

物体をある高さから水平方向に初速度を与えると放物線を描いて飛ぶ。



```julia
g = 9.8
vx0=20
ts = range(0,10,step=0.1)
ys = 1/2 * g.*ts.^2
xs = vx0.*ts
vs= round.(g.*ts,digits=3)
theta = range(0,2π,step=0.01)
r = 5 # 質点の半径
anim = Animation()
for i in 1:length(ts)
    plt=plot(xs[i] .+r*cos.(theta),ys[i].+r*sin.(theta),aspectratio=1,xlims=(-10,100),ylims=(-10,100),yaxis=:flip,xlabel="x",ylabel="y",linewidth=5)
    frame(anim,plt)
end
gif(anim,fps=15)
```

$x,y$方向の運動を別々に観察してみよう。

```julia
g = 9.8
vx0=20
ts = range(0,10,step=0.1)
ys = 1/2 * g.*ts.^2
xs = vx0.*ts
vs= round.(g.*ts,digits=3)
theta = range(0,2π,step=0.01)
r = 5 # 質点の半径
anim = Animation()
for i in 1:length(ts)
    plt=plot(xs[i] .+r*cos.(theta),ys[i].+r*sin.(theta),aspectratio=1,xlims=(-10,100),ylims=(-10,100),yaxis=:flip,label="",linewidth=5)
    plt=plot!(xs[1:i],ys[1:i],label="",linewidth=5)
    plt=plot!(0 .+r*cos.(theta),ys[i].+r*sin.(theta),label="",linewidth=5)
    plt=plot!(xs[i] .+r*cos.(theta),0 .+r*sin.(theta),label="",linewidth=5)
    frame(anim,plt)
end
gif(anim,fps=20)
```

水平投射は、鉛直方向は自由落下、水平方向は等速直線運動をしている。
水平方向の等速直線運動の速度は、水平方向の初速度で決まる。

## 斜方投射

### 斜方投射の軌道
物体を斜め上方に投げ上げると、物体は放物線を描いて上昇し、やがて落下に転じる。

斜方投射の運動の様子を見てみよう。

```julia
g = 9.8
# vx0=10
# vy0=30
v0=35
θ=π/2.5
vx0=v0*cos(θ)
vy0=v0*sin(θ)
ts = range(0,10,step=0.1)
xs = vx0.*ts
ys = -1/2 * g.*ts.^2 .+vy0.*ts 
vs= round.(g.*ts,digits=3)
theta = range(0,2π,step=0.01)
r = 5 # 質点の半径
anim = Animation()
for i in 1:length(ts)
    plt=plot(xs[i] .+r*cos.(theta),ys[i].+r*sin.(theta),aspectratio=1,xlims=(-10,100),ylims=(-10,100),label="",linewidth=5)
    plt=plot!(xs[1:i],ys[1:i],label="",xlabel="x",ylabel="y",linewidth=5)
    frame(anim,plt)
end
gif(anim,fps=20)
```

水平投射の場合と同様に、$x,y$方向の運動を別々に観察したものが次の図である。

```julia
g = 9.8
v0=35
θ=π/2.5
vx0=v0*cos(θ)
vy0=v0*sin(θ)
ts = range(0,10,step=0.1)
xs = vx0.*ts
ys = -1/2 * g.*ts.^2 .+vy0.*ts 
vs= round.(g.*ts,digits=3)
theta = range(0,2π,step=0.01)
r = 5 # 質点の半径
anim = Animation()
for i in 1:length(ts)
    plt=plot(xs[i] .+r*cos.(theta),ys[i].+r*sin.(theta),aspectratio=1,xlims=(-10,100),ylims=(-10,100),label="",linewidth=5)
    plt=plot!(xs[1:i],ys[1:i],label="",xlabel="x",ylabel="y",linewidth=5)
    plt=plot!(0 .+r*cos.(theta),ys[i].+r*sin.(theta),label="",linewidth=5)
    plt=plot!(xs[i] .+r*cos.(theta),0 .+r*sin.(theta),label="",linewidth=5)
    frame(anim,plt)
end
gif(anim,fps=20)
```

上の図から、斜方投射について次のことがわかる。

1. 鉛直方向の運動は垂直投げ上げ
2. 水平方向の運動は等速直線運動
3. 運動の軌道は上に凸の放物線



### 斜方投射の式

次に、斜方投射の運動が式でどのように表されるかを考える。
いま、小球を

- 速さ$v_0(>0)$[m/s]　で
- 水平方向と$\theta$をなすむきに
  
投げたとする。
そして、水平方向に$x$軸、垂直方向に$y$軸を取る。




投げた瞬間の時刻を$t=0$とし、ある時刻$t=t$での

- 速度の$x$成分を$v_x$[m/s]
- 速度の$y$成分を$v_y$[m/s]
- 小球の座標を$(x,y)$

とする。

まず、$x$軸方向の運動から考える。
これは等速直線運動であることがすでにわかっている。
その初速度は、

```julia
v0=30
θ = π/3
vx=v0*cos(θ)
vy=v0*sin(θ)
plot([0,vx], [0,vy], arrow = arrow(),xlabel="x",ylabel="y",xlims=(-5,30),ylims=(-5,30),aspectratio=1,label="v_0",linewidth=5)
plot!([0,vx], [0,0], arrow = arrow(),label="vx_0",linewidth=5)
plot!([0,0], [0,vy], arrow = arrow(),label="vy_0",linewidth=5)
annotate!(3, 2, ("θ",20))
```

より、$$v_x = v_0 \cos{\theta}$$であり、このことから変位は
$$x = v_0 \cos{\theta}\cdot t$$
とかける。

次に$y$方向の運動について考える。
$y$方向の運動は、初速度$v_0\sin{\theta}$の鉛直投げ上げ運動とみなせるのだったから、
$$v_y=v_0\cos{\theta}-gt$$
$$y = v_0\sin{\theta}\cdot t - \frac{1}{2}gt^2$$
のように速度と変位は表せる。

ここまでで得られた$x,y$方向の変位の式
$$x = v_0 \cos{\theta}\cdot t$$
$$y = v_0\sin{\theta}\cdot t - \frac{1}{2}gt^2$$
から$t$を消去すると、
$$y = \tan{\theta} \cdot x - \frac{g}{{2v_0}^2\cos^2{\theta}}\cdot x^2$$
が得られる。
$x,y$以外のごちゃごちゃしたものをシンプルな文字で置き直せば、
$$y = \alpha  x - \beta x^2 $$
とみなすことができ、$\beta>0$であるのでこれは小球の運動が上に凸の放物線になることを表している。

### いくつかの注意

斜方投射と鉛直投げ上げ運動にはともに「最高点」がある。
これは鉛直方向の速度が$0$になり、正から負に切り替わる点である。

鉛直投げ上げの場合は鉛直方向のみに運動しているため、最高点では運動自体の速度も$0$になる。
一方、斜方投射の場合は水平方向の等速直線運動があるため、運動自体の速度は$0$にはならない。

```julia
g = 9.8
v0=35
θ=π/2.5
vx0=v0*cos(θ)
vy0=v0*sin(θ)
ts = range(0,10,step=0.1)
xs = vx0.*ts
ys = -1/2 .* g.*ts.^2 .+vy0.*ts 
xvs= vx0.+0 .*ts
yvs=vy0 .-g .*ts
theta = range(0,2π,step=0.01)
r = 5 # 質点の半径
anim = Animation()
for i in 1:length(ts)
    plt=plot(xs[i] .+r*cos.(theta),ys[i].+r*sin.(theta),aspectratio=1,xlims=(-10,100),ylims=(-10,100),label="",xlabel="x",ylabel="y",linewidth=5)
    #plt=plot!(xs[1:i],ys[1:i],label="")
    plt=plot!([xs[i],xs[i]+xvs[i]], [ys[i],ys[i]+yvs[i]], arrow = arrow(),label="v",linewidth=5)
    plt=plot!([xs[i],xs[i]], [ys[i],ys[i]+yvs[i]], arrow = arrow(),label="vy",linewidth=5)
    plt=plot!([xs[i],xs[i]+xvs[i]], [ys[i],ys[i]], arrow = arrow(),label="vx",linewidth=5)
    frame(anim,plt)
end
gif(anim,fps=20)
```

上の図はそれぞれの瞬間の速度を矢印で表したものである。
最高到達点では確かに$y$方向の速度が$0$になってその後向きが反転する。
一方で$x$方向は等速直線運動しているため、初速度のまま変わらない。
