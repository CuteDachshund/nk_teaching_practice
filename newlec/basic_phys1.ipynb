```julia
using Plots
```

```julia
g = 9.8
ts = range(0,5,step=1/30)
ys = 1/2 * g.*ts.^2
vs= round.(g.*ts,digits=3)
theta = range(0,2π,step=0.01)
r = 5 # 質点の半径
anim = Animation()
for i in 1:length(ts)
    t = round(ts[i],digits=3)
    plt=plot(0 .+r*cos.(theta),ys[i].+r*sin.(theta),aspectratio=1,xlims=(-50,50),ylims=(-10,100),yaxis=:flip,label="t = $(t)",ylabel="y",linewidth=2,linecolor=:blue)
    frame(anim,plt)
end
gif(anim,"freefall.gif",fps=30)
```

```julia
g = 9.8
ts = range(0,5,step=1/10)
ys = 1/2 * g.*ts.^2
vs= round.(g.*ts,digits=3)
theta = range(0,2π,step=0.01)
r = 5 # 質点の半径
anim = Animation()
plt=plot()
for i in 1:length(ts)
    t = round(ts[i],digits=3)
    plt=plot!(0 .+r*cos.(theta),ys[i].+r*sin.(theta),aspectratio=1,xlims=(-50,50),ylims=(-10,100),yaxis=:flip,label="",ylabel="y",linewidth=2,linecolor=:blue)
    frame(anim,plt)
end
gif(anim,"freefallstrobo.gif",fps=10)
```
